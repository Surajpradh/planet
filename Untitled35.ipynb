## Define System Constants and Parameters

### Subtask:
Define physical constants (e.g., elementary charge `q_s`, particle mass `m_s`, Boltzmann constant `k_B`, speed of light `c`, reference wavelength `lambda_0`) and system-specific parameters like loop length `L`. Also, define placeholder functions for equilibrium quantities such as `B_0(s)`, `n_0(s)`, and `T_0(s)`.


**Reasoning**:
The subtask requires defining physical constants, system-specific parameters, and placeholder functions. This can be achieved by importing `numpy` and then defining these variables and functions within a single code block.



```python
import numpy as np

# Define physical constants
q_s = 1.602e-19  # Elementary charge (Coulombs) - assuming proton charge
m_s = 1.672e-27  # Particle mass (kg) - assuming proton mass
k_B = 1.380e-23  # Boltzmann constant (J/K)
c = 2.998e8    # Speed of light (m/s)
lambda_0 = 103.0e-9 # Reference wavelength (m) - e.g., Lyman-beta for hydrogen

# Define system-specific parameter
L = 1.0e7        # Loop length (meters) - example value

# Placeholder functions for equilibrium quantities
def B_0(s):
    """Placeholder function for equilibrium magnetic field."""
    # For now, return a constant value. In a real scenario, this would be a function of s.
    return 0.1 # Tesla

def n_0(s):
    """Placeholder function for equilibrium number density."""
    # For now, return a constant value. In a real scenario, this would be a function of s.
    return 1.0e15 # particles/m^3

def T_0(s):
    """Placeholder function for equilibrium temperature."""
    # For now, return a constant value. In a real scenario, this would be a function of s.
    return 1.0e6 # Kelvin

print("Physical constants, system parameter, and placeholder functions defined.")
```

## Setup Computational Grids

### Subtask:
Establish numerical grids for the spatial coordinate `s` and the velocity components `v_parallel` and `v_perp` using NumPy arrays. These grids will be used for numerical integration and evaluation of functions.


**Reasoning**:
The subtask requires setting up computational grids for spatial coordinate `s` and velocity components `v_parallel` and `v_perp`. This involves defining the number of points, calculating a thermal velocity for scaling, and using `np.linspace` to create these grids. Finally, the shapes of the grids need to be printed for verification.



```python
import numpy as np

# 1. Define the number of points for the spatial and velocity grids
s_points = 100
v_points = 200

# 2. Create the spatial grid `s_grid`
s_grid = np.linspace(0, L, s_points)

# 3. Create the parallel velocity grid `v_parallel_grid`
# Estimate characteristic thermal velocity using T_0(0)
v_thermal = np.sqrt(2 * k_B * T_0(0) / m_s)
v_parallel_grid = np.linspace(-5 * v_thermal, 5 * v_thermal, v_points)

# 4. Create the perpendicular velocity grid `v_perp_grid`
v_perp_grid = np.linspace(0, 5 * v_thermal, v_points)

# 5. Print the shapes of the created grids to verify their dimensions
print(f"Shape of s_grid: {s_grid.shape}")
print(f"Shape of v_parallel_grid: {v_parallel_grid.shape}")
print(f"Shape of v_perp_grid: {v_perp_grid.shape}")

print("Computational grids s, v_parallel, and v_perp defined.")
```

## Define Simplified Perturbed Quantities

### Subtask:
Define simplified analytical forms for the perturbed distribution function `f_s1(v, s, t)`, electric field `E1(s, t)`, and magnetic field `B1(s, t)`. These simplified forms will allow for the computation of observables.


**Reasoning**:
The subtask requires defining simplified analytical forms for the perturbed distribution function `f_s1`, electric field `E1`, and magnetic field `B1`. This can be achieved by creating placeholder Python functions using `numpy` for sinusoidal variations and a Maxwellian-like term for `f_s1`.



```python
import numpy as np

# Placeholder functions for perturbed quantities

def f_s1(v_parallel, v_perp, s, t=0):
    """Placeholder function for the perturbed distribution function f_s1.

    For simplicity, it's a sinusoidal perturbation in 's' multiplied by a
    Maxwellian-like velocity dependence. The time dependence is ignored (t=0).
    """
    # Example: A small perturbation proportional to a Maxwellian and a sine wave in s.
    # Using a characteristic thermal velocity for scaling the Maxwellian part.
    # v_thermal is already defined in a previous cell.

    # Amplitude of the perturbation
    amplitude = 1e10 # example amplitude, chosen to be small relative to n_0

    # Spatial variation (sinusoidal along the loop, assuming some wavelength related to L)
    k_s = 2 * np.pi / L # Wavenumber for spatial variation
    spatial_variation = np.sin(k_s * s)

    # Velocity dependence (Maxwellian-like term for simplicity)
    # Assuming an effective temperature for the perturbation, here using T_0(s)
    T_eff = T_0(s) # Using the equilibrium temperature as a reference
    v_th_sq = 2 * k_B * T_eff / m_s

    # Combine parallel and perpendicular velocities in an effective speed squared for simplicity
    # This is a simplification and would be more complex in a full kinetic treatment
    v_sq = v_parallel**2 + v_perp**2
    velocity_distribution = np.exp(-v_sq / v_th_sq)

    return amplitude * spatial_variation * velocity_distribution

def E1(s, t=0):
    """Placeholder function for the perturbed electric field E1.

    For simplicity, it's a sinusoidal function of 's' with a fixed amplitude.
    The time dependence is ignored (t=0).
    """
    # Amplitude of the perturbed electric field
    E1_amplitude = 1e-4 # Volts/meter (example value)

    # Spatial variation (sinusoidal along the loop)
    k_s = 2 * np.pi / L # Wavenumber for spatial variation
    spatial_variation = np.sin(k_s * s)

    return E1_amplitude * spatial_variation

def B1(s, t=0):
    """Placeholder function for the perturbed magnetic field B1.

    For simplicity, it's a sinusoidal function of 's' with a fixed amplitude.
    The time dependence is ignored (t=0).
    """
    # Amplitude of the perturbed magnetic field
    B1_amplitude = 1e-8 # Tesla (example value)

    # Spatial variation (sinusoidal along the loop)
    k_s = 2 * np.pi / L # Wavenumber for spatial variation
    spatial_variation = np.sin(k_s * s + np.pi/2) # Phase shift for B1 relative to E1 if desired

    return B1_amplitude * spatial_variation

print("Simplified analytical forms for f_s1, E1, and B1 defined.")
```

## Implement Observables Calculation

### Subtask:
Write Python functions to calculate the specified observables: heating `Q(s)`, parallel and perpendicular temperatures `T_parallel` and `T_perp`, and spectral line broadening `Delta_lambda_squared`. These functions will utilize the defined grids, equilibrium parameters, and the simplified perturbed quantities.


**Reasoning**:
The subtask requires implementing several Python functions to calculate observables, including heating, parallel and perpendicular temperatures, and spectral line broadening. This involves calculating step sizes, defining the equilibrium distribution function, creating a general velocity space integration helper, and then implementing each specific observable function. Finally, these functions will be called in a loop over the spatial grid to gather results.



```python
import numpy as np

# 1. Calculate the step sizes dv_parallel and dv_perp
dv_parallel = v_parallel_grid[1] - v_parallel_grid[0]
dv_perp = v_perp_grid[1] - v_perp_grid[0]

# 2. Define f_s0(v_parallel, v_perp, s)
def f_s0(v_parallel, v_perp, s_val):
    """Returns the equilibrium Maxwell-Boltzmann distribution function."""
    n_0_val = n_0(s_val)
    T_0_val = T_0(s_val)

    # Ensure T_0_val is not zero to avoid division by zero
    if T_0_val <= 0:
        return np.zeros_like(v_parallel) * np.zeros_like(v_perp) # Return zero for non-physical temperatures

    coeff = n_0_val * (m_s / (2 * np.pi * k_B * T_0_val))**(3/2)
    exponent = -m_s * (v_parallel**2 + v_perp**2) / (2 * k_B * T_0_val)
    return coeff * np.exp(exponent)

# 3. Define a helper function to integrate over velocity space
def integrate_velocity_space(integrand_func, s_val, V_PARALLEL_2D, V_PERP_2D):
    """Performs 2D numerical integration over the v_parallel_grid and v_perp_grid."""
    # The integrand_func should take V_PARALLEL_2D, V_PERP_2D, and s_val as arguments.
    # It should return a 2D array of the integrand evaluated at each point.
    integrand_evaluated = integrand_func(V_PARALLEL_2D, V_PERP_2D, s_val)

    # Calculate the volume element in cylindrical coordinates (2 * pi * v_perp * dv_parallel * dv_perp)
    # The integration is for f * dv_parallel * v_perp * dv_perp, so the 2*pi is usually a constant
    # or included when transforming from spherical/cylindrical to Cartesian velocity space integration.
    # Here we assume a cylindrical integral (v_perp is radial component).
    # The factor of 2*pi is often included for integrals of f(v_parallel, v_perp) * v_perp dv_perp dv_parallel.
    # In the context of `integrate_velocity_space` as a helper for calculating moments, it typically represents
    # Integral(integrand * dv_parallel * dv_perp) * (2 * np.pi * v_perp)
    # If integrand_evaluated already incorporates f, then we need to include the v_perp factor and the 2*pi constant.

    # Check if integrand_evaluated is already a 2D array, if not, it means the integrand_func already applied the meshgrid.
    # For the Maxwellian and f_s1, they take v_parallel and v_perp as arrays and return an array directly.
    # If integrand_func returns a scalar, it's an error. It should return a 2D array compatible with V_PARALLEL_2D/V_PERP_2D.

    # Sum over the 2D grid, multiplying by the velocity space volume element.
    # For cylindrical coordinates, the volume element is v_perp * dv_perp * dv_parallel * d(phi) = v_perp * dv_perp * dv_parallel * 2*pi
    # We need to reshape V_PERP_2D to match the integrand_evaluated if it's not already aligned.
    # Ensure V_PERP_2D is correctly passed and used for the volume element.

    # When performing the meshgrid, the resulting 2D arrays are (len(v_perp_grid), len(v_parallel_grid))
    # So, V_PERP_2D contains the v_perp values for each row, and V_PARALLEL_2D for each column.

    # The integration is sum( f(v_parallel, v_perp) * 2 * pi * v_perp * dv_parallel * dv_perp )
    # The `integrand_func` must return the `f(v_parallel, v_perp)` part.

    # Make sure v_perp is broadcastable to the shape of integrand_evaluated
    # V_PERP_2D is already a 2D array, use it directly.

    integration_result = np.sum(integrand_evaluated * V_PERP_2D * dv_parallel * dv_perp) * (2 * np.pi)
    return integration_result

# Create 2D meshgrids for velocity components once
V_PARALLEL_2D, V_PERP_2D = np.meshgrid(v_parallel_grid, v_perp_grid, indexing='xy')

# 4. Implement calculate_Q(s_val)
def calculate_Q(s_val):
    """Calculates the heating Q(s_val)."""
    # The integrand for Q is (v_parallel * f_s1).
    # f_s1 takes v_parallel and v_perp as (2D) arrays
    def integrand_Q(V_PARALLEL, V_PERP, s):
        # f_s1 is defined to handle array inputs directly
        return V_PARALLEL * f_s1(V_PARALLEL, V_PERP, s)

    Q = -q_s * E1(s_val) * integrate_velocity_space(integrand_Q, s_val, V_PARALLEL_2D, V_PERP_2D)
    return Q

# 5. Implement calculate_temperatures(s_val)
def calculate_temperatures(s_val):
    """Calculates n_total, T_parallel, and T_perp for a given s_val."""
    # Calculate total distribution function f_total = f_s0 + f_s1
    f_total_2d = f_s0(V_PARALLEL_2D, V_PERP_2D, s_val) + f_s1(V_PARALLEL_2D, V_PERP_2D, s_val)

    # For calculating moments, the integrand_func for integrate_velocity_space will simply be the f_total_2d itself
    # or f_total_2d multiplied by the appropriate velocity component squared.
    # The helper function needs to be flexible, or we define specific integrands.

    # Function for density integration (integrand is just f_total_2d)
    def integrand_n_total(V_PARALLEL, V_PERP, s):
        # f_total_2d is already calculated outside for efficiency
        return f_total_2d

    n_total = integrate_velocity_space(integrand_n_total, s_val, V_PARALLEL_2D, V_PERP_2D)

    # Function for T_parallel integration (integrand is v_parallel^2 * f_total_2d)
    def integrand_T_parallel(V_PARALLEL, V_PERP, s):
        return V_PARALLEL**2 * f_total_2d

    # Function for T_perp integration (integrand is v_perp^2 * f_total_2d)
    def integrand_T_perp(V_PARALLEL, V_PERP, s):
        return V_PERP**2 * f_total_2d

    # Calculate moments
    if n_total > 0: # Avoid division by zero
        T_parallel = (m_s / (k_B * n_total)) * integrate_velocity_space(integrand_T_parallel, s_val, V_PARALLEL_2D, V_PERP_2D)
        T_perp = (m_s / (2 * k_B * n_total)) * integrate_velocity_space(integrand_T_perp, s_val, V_PARALLEL_2D, V_PERP_2D)
    else:
        T_parallel = 0.0 # Or some appropriate default/error value
        T_perp = 0.0

    return n_total, T_parallel, T_perp

# 6. Implement calculate_line_broadening(T_parallel_val)
def calculate_line_broadening(T_parallel_val):
    """Calculates spectral line broadening Delta_lambda_squared."""
    # For a purely thermal broadening, it depends on T_parallel.
    Delta_lambda_squared = (lambda_0 / c)**2 * (2 * k_B * T_parallel_val / m_s)
    return Delta_lambda_squared

# 7. Initialize empty lists to store the results
Q_s_values = []
n_total_values = []
T_parallel_values = []
T_perp_values = []
delta_lambda_squared_values = []

# 8. Loop through each s_val in s_grid and calculate observables
print("Calculating observables for each point in s_grid...")
for i, s_val in enumerate(s_grid):
    # Calculate Q(s_val)
    Q_val = calculate_Q(s_val)
    Q_s_values.append(Q_val)

    # Calculate temperatures and density
    n_tot, T_par, T_per = calculate_temperatures(s_val)
    n_total_values.append(n_tot)
    T_parallel_values.append(T_par)
    T_perp_values.append(T_per)

    # Calculate spectral line broadening
    delta_lambda_sq = calculate_line_broadening(T_par)
    delta_lambda_squared_values.append(delta_lambda_sq)

    if (i + 1) % 10 == 0 or i == len(s_grid) - 1:
        print(f"  Processed {i+1}/{len(s_grid)} spatial points.")

# 9. Print a message indicating completion
print("Observable calculations complete.")

# Convert lists to numpy arrays for easier plotting/analysis later
Q_s_values = np.array(Q_s_values)
n_total_values = np.array(n_total_values)
T_parallel_values = np.array(T_parallel_values)
T_perp_values = np.array(T_perp_values)
delta_lambda_squared_values = np.array(delta_lambda_squared_values)
```

## Perform Parameter Study and Visualize Results

### Subtask:
Conduct a parameter study by varying the equilibrium magnetic field `B_0`, calculating heating `Q(s)` and temperature anisotropy `T_parallel/T_perp` for each variation. Then, generate and display plots of `Q(s)` vs. `s` and `T_parallel/T_perp` vs. `s` for all `B_0` values, including appropriate labels and legends.


**Reasoning**:

```python
import matplotlib.pyplot as plt

# 1. Define a list of B_0 values for the parameter study
B0_study_values = [0.05, 0.1, 0.2] # in Tesla

# 2. Create empty dictionaries to store the results
all_Q_s_values = {}
all_T_ratio_values = {}

# 3. Store the original B_0 function
original_B0_func = B_0

print("Starting parameter study for B_0...")

# 4. Start a loop that iterates through each B0_val
for B0_val in B0_study_values:
    print(f"  Processing B_0 = {B0_val} T")

    # 5. Temporarily redefine the global B_0 function
    def B_0(s_placeholder):
        return B0_val

    # 6. Initialize empty lists for current B0_val
    current_Q_s = []
    current_T_parallel = []
    current_T_perp = []

    # 7. Loop through each s_val in the s_grid to recalculate observables
    for s_val in s_grid:
        # Calculate Q(s_val)
        Q_val = calculate_Q(s_val)
        current_Q_s.append(Q_val)

        # Calculate temperatures and density
        _, T_par, T_per = calculate_temperatures(s_val)
        current_T_parallel.append(T_par)
        current_T_perp.append(T_per)

    # 8. Convert lists to NumPy arrays
    current_Q_s_array = np.array(current_Q_s)
    current_T_parallel_array = np.array(current_T_parallel)
    current_T_perp_array = np.array(current_T_perp)

    # 9. Calculate the temperature ratio T_parallel / T_perp
    # Handle potential division by zero or very small T_perp_values
    T_ratio_array = np.divide(current_T_parallel_array, current_T_perp_array,
                              out=np.zeros_like(current_T_parallel_array),
                              where=current_T_perp_array!=0)
    # Replace potential inf/nan if any arise from other issues (e.g., T_parallel also zero where T_perp is zero)
    T_ratio_array[np.isinf(T_ratio_array)] = np.nan # Or a large number, depending on desired behavior
    T_ratio_array[np.isnan(T_ratio_array)] = 0.0    # Or some default value if both are zero

    # 10. Store the results in the dictionaries
    all_Q_s_values[B0_val] = current_Q_s_array
    all_T_ratio_values[B0_val] = T_ratio_array

print("Parameter study complete.")

# 11. Restore the original B_0 function
B_0 = original_B0_func
print("Original B_0 function restored.")

# 12. matplotlib.pyplot is already imported as plt

# 13. Create a figure with two subplots
fig, axes = plt.subplots(2, 1, figsize=(10, 12), sharex=True)

# 14. Plot Q(s) vs. s
ax1 = axes[0]
for B0_val in B0_study_values:
    ax1.plot(s_grid, all_Q_s_values[B0_val], label=f'B0 = {B0_val} T')
ax1.set_ylabel('Heating Rate Q (W/m^3)')
ax1.set_title('Heating Profile for Different $B_0$')
ax1.legend()
ax1.grid(True)

# 15. Plot T_parallel/T_perp vs. s
ax2 = axes[1]
for B0_val in B0_study_values:
    ax2.plot(s_grid, all_T_ratio_values[B0_val], label=f'B0 = {B0_val} T')
ax2.set_xlabel('Spatial Coordinate s (m)')
ax2.set_ylabel('Temperature Ratio $T_{\parallel} / T_{\perp}$ ')
ax2.set_title('Temperature Anisotropy for Different $B_0$')
ax2.legend()
ax2.grid(True)

# 16. Adjust layout and display plots
plt.tight_layout()
plt.show()

print("Plots generated for parameter study.")
```

```python
import numpy as np

# --- Define nu_damping ---
# Damping rate for Landau resonance term (example value, could be more physically derived)
# This is a phenomenological damping factor to prevent singularities at resonance
nu_damping = 1e-3 # rad/s

# Analytical derivatives for equilibrium quantities
# These rely on global parameters defined in previous cells (e.g., L, n_0_min, B_0_min etc.)

def dn_0_ds(s):
    """Analytical derivative of n_0(s) with respect to s."""
    global n_0_min, n_0_base, L
    return (n_0_min - n_0_base) / L

def dT_0_ds(s):
    """Analytical derivative of T_0(s) with respect to s."""
    global T_0_amplitude, L
    # T_0(s) = T_0_base + T_0_amplitude * np.sin(np.pi * s / L)
    return T_0_amplitude * (np.pi / L) * np.cos(np.pi * s / L)

def dB_0_ds(s):
    """Analytical derivative of B_0(s) with respect to s."""
    global B_0_min, B_0_max, L
    # B_0(s) = A*s**2 + B_coeff*s + C
    A = 4 * (B_0_min - B_0_max) / (L**2)
    B_coeff = -A * L
    return 2 * A * s + B_coeff

# Helper function for the velocity part of f_s0 (without n_0 scaling)
def f_s0_velocity_part(v_parallel, v_perp, T_0_val):
    """Returns the velocity-dependent part of the equilibrium Maxwellian (normalized to 1 in velocity space)."""
    global m_s, k_B
    if T_0_val <= 0:
        return np.zeros_like(v_parallel) * np.zeros_like(v_perp)
    coeff = (m_s / (2 * np.pi * k_B * T_0_val))**(3/2)
    exponent = -m_s * (v_parallel**2 + v_perp**2) / (2 * k_B * T_0_val)
    return coeff * np.exp(exponent)

# --- GLOBAL SCALING FACTORS for perturbed quantities ---
# These are now global so they can be varied in parameter studies
E1_amplitude_scaling_factor = 5e-4 # Example scaling factor (arbitrary, adjust as needed)
B1_amplitude_scaling_factor = 5e-9 # Example scaling factor (arbitrary, adjust as needed)

# Coefficients for various effects (phenomenological scaling factors to keep f_s1 a small perturbation)
E1_effect_scale = 1e-15    # Scales the electric field driven perturbation
B1_effect_scale = 1e-15    # Scales the magnetic field driven perturbation
grad_T_effect_scale = 1e-15  # Scales the temperature gradient driven perturbation
grad_n_effect_scale = 1e-15  # Scales the density gradient driven perturbation
grad_B_effect_scale = 1e-15  # Scales the magnetic field gradient driven perturbation

def E1(s, t=0):
    """Modified placeholder function for the perturbed electric field E1.
    Now scales with B_0(s) and uses wave parameters (omega, k_parallel_val).
    """
    global L, omega, k_parallel_val, E1_amplitude_scaling_factor
    B_0_val = B_0(s)

    # Amplitude of the perturbed electric field now scales with B_0(s)
    # The scaling factor is adjusted to ensure E1 is a perturbation
    E1_amplitude = E1_amplitude_scaling_factor * B_0_val

    # Spatial variation (sinusoidal along the loop, for t=0)
    spatial_variation = np.sin(k_parallel_val * s)

    return E1_amplitude * spatial_variation

def B1(s, t=0):
    """Modified placeholder function for the perturbed magnetic field B1.
    Now scales with B_0(s) and uses wave parameters (omega, k_parallel_val).
    """
    global L, omega, k_parallel_val, B1_amplitude_scaling_factor
    B_0_val = B_0(s)

    # Amplitude of the perturbed magnetic field now scales with B_0(s)
    B1_amplitude = B1_amplitude_scaling_factor * B_0_val

    # Spatial variation (sinusoidal along the loop, with a phase shift relative to E1)
    spatial_variation = np.sin(k_parallel_val * s + np.pi/2) # Phase shift for B1 relative to E1

    return B1_amplitude * spatial_variation

def f_s1(v_parallel, v_perp, s, t=0):
    """Modified placeholder function for the perturbed distribution function f_s1.
    It explicitly depends on equilibrium quantities, their gradients, E1, B1,
    and incorporates a simplified Landau resonance term and other kinetic effects.
    """
    global m_s, k_B, L, omega, k_parallel_val, nu_damping
    global E1_effect_scale, B1_effect_scale, grad_T_effect_scale, grad_n_effect_scale, grad_B_effect_scale

    n_0_val = n_0(s)
    T_0_val = T_0(s)
    B_0_val = B_0(s)

    # Derivatives of equilibrium quantities
    dn_0_ds_val = dn_0_ds(s)
    dT_0_ds_val = dT_0_ds(s)
    dB_0_ds_val = dB_0_ds(s)

    # Ensure T_0_val, n_0_val, B_0_val are positive and non-zero to avoid division errors
    T_0_safe = np.maximum(T_0_val, 1e3)
    n_0_safe = np.maximum(n_0_val, 1e10)
    B_0_safe = np.maximum(B_0_val, 1e-5)

    # The velocity part of the equilibrium Maxwellian (normalized to 1 in velocity space)
    f_s0_vel_part = f_s0_velocity_part(v_parallel, v_perp, T_0_safe)

    # Calculate local thermal velocity squared for scaling
    v_th_sq = np.maximum(2 * k_B * T_0_safe / m_s, 1e-10)
    v_thermal = np.sqrt(v_th_sq)

    # Get perturbed E1 and B1 values
    E1_val = E1(s, t)
    B1_val = B1(s, t)

    # Initialize dimensionless perturbation factor for f_s1
    # f_s1 = n_0(s) * f_s0_velocity_part * dimensionless_perturbation_factor
    dimensionless_perturbation_factor = np.zeros_like(v_parallel) + np.zeros_like(v_perp)

    # 1. E1-driven Landau-like kinetic term
    # Represents response to E1 with a resonance around v_parallel = omega/k_parallel_val
    # Using a dimensionless form of the dispersive Landau response profile
    resonance_factor_E1 = omega * (omega - k_parallel_val * v_parallel) / ((omega - k_parallel_val * v_parallel)**2 + nu_damping**2)
    # Scale E1_val to be dimensionless (e.g., E1 / (B_0 * v_thermal))
    E1_dimensionless = E1_val / (B_0_safe * v_thermal + 1e-15) # Add small number to prevent division by zero

    dimensionless_perturbation_factor += E1_effect_scale * E1_dimensionless * (v_parallel / v_thermal) * resonance_factor_E1

    # 2. B1-driven term (e.g., related to magnetic mirroring/compression)
    # Scale B1_val to be dimensionless (B1 / B_0)
    B1_dimensionless = B1_val / (B_0_safe + 1e-15) # Add small number to prevent division by zero

    dimensionless_perturbation_factor += B1_effect_scale * B1_dimensionless * (v_perp / v_thermal)**2

    # 3. Gradient-driven terms (thermal, density, magnetic gradients)
    # These terms are made dimensionless by multiplying by L (loop length)
    # and using ratios of derivatives to equilibrium quantities.
    dimensionless_perturbation_factor += grad_n_effect_scale * (dn_0_ds_val / n_0_safe) * L * (v_parallel / v_thermal)**2
    dimensionless_perturbation_factor += grad_T_effect_scale * (dT_0_ds_val / T_0_safe) * L * (v_parallel / v_thermal)**2
    dimensionless_perturbation_factor += grad_B_effect_scale * (dB_0_ds_val / B_0_safe) * L * (v_perp / v_thermal)**2

    # The total perturbed distribution function
    # This ensures f_s1 has the same units as f_s0 (particles / m^3 / (m/s)^3 = s^3/m^6)
    return n_0_val * f_s0_vel_part * dimensionless_perturbation_factor

print("Modified analytical forms for f_s1, E1, and B1 defined with enhanced physical dependencies.")

# --- Parameter Study for E1_amplitude_scaling_factor ---

# 1. Define a list of E1_amplitude_scaling_factor values to study
E1_amplitude_scaling_factor_study_values = [1e-4, 5e-4, 1e-3]

# 2. Store the current global E1_amplitude_scaling_factor to restore it after the study.
original_E1_amplitude_scaling_factor = E1_amplitude_scaling_factor

# 3. Create empty dictionaries to store the calculated Q(s) and T_parallel/T_perp arrays
all_Q_s_values_E1_scale = {}
all_T_ratio_values_E1_scale = {}

print("\nStarting parameter study for E1_amplitude_scaling_factor...")

# 4. Start a loop that iterates through each defined E1_amplitude_scaling_factor value.
for E1_scale_val in E1_amplitude_scaling_factor_study_values:
    print(f"  Processing E1_amplitude_scaling_factor = {E1_scale_val}")

    # 5. Temporarily set the global E1_amplitude_scaling_factor to the current value.
    E1_amplitude_scaling_factor = E1_scale_val

    # 6. Initialize empty lists for the results of the current E1_amplitude_scaling_factor.
    current_Q_s = []
    current_T_parallel = []
    current_T_perp = []

    # 7. Loop through each spatial point s_val in s_grid:
    for i, s_val in enumerate(s_grid):
        # a. Call calculate_Q(s_val) and append the result to current_Q_s.
        Q_val = calculate_Q(s_val)
        current_Q_s.append(Q_val)

        # b. Call calculate_temperatures(s_val) to get n_total, T_parallel, and T_perp.
        _, T_par, T_per = calculate_temperatures(s_val)
        current_T_parallel.append(T_par)
        current_T_perp.append(T_per)

    # 8. Convert current_Q_s, current_T_parallel, and current_T_perp lists into NumPy arrays.
    current_Q_s_array = np.array(current_Q_s)
    current_T_parallel_array = np.array(current_T_parallel)
    current_T_perp_array = np.array(current_T_perp)

    # 9. Calculate the temperature ratio T_parallel / T_perp for the current scaling factor.
    T_ratio_array = np.divide(current_T_parallel_array, current_T_perp_array,
                              out=np.zeros_like(current_T_parallel_array),
                              where=current_T_perp_array!=0)
    T_ratio_array[np.isinf(T_ratio_array)] = np.nan
    T_ratio_array[np.isnan(T_ratio_array)] = 0.0

    # 10. Store the resulting Q(s) array and T_parallel/T_perp ratio array in the dictionaries.
    all_Q_s_values_E1_scale[E1_scale_val] = current_Q_s_array
    all_T_ratio_values_E1_scale[E1_scale_val] = T_ratio_array

print("Parameter study for E1_amplitude_scaling_factor complete.")

# 11. After the loop completes, restore the global E1_amplitude_scaling_factor to its original value.
E1_amplitude_scaling_factor = original_E1_amplitude_scaling_factor
print("Original E1_amplitude_scaling_factor restored.")
```
